## Hotel Aggregator Platform Analytics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("hotel_aggregator_dataset.csv")
data.head()
```

```python
data.drop(['listing_url','listing_url','scrape_id','source','name','description','neighborhood_overview','picture_url','host_url','host_location','host_about','host_thumbnail_url','host_picture_url','neighbourhood','neighbourhood_group_cleansed','amenities','minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights','maximum_maximum_nights','calendar_updated','availability_60','availability_90','availability_365','calendar_last_scraped','host_neighbourhood','number_of_reviews_l30d','first_review','last_review','review_scores_accuracy','review_scores_value','license','calculated_host_listings_count','calculated_host_listings_count_entire_homes','calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms','bathrooms'],inplace=True,axis=1)

data.head()
```

```python
print(data.isnull().sum())
```

```python
data.dropna(how='any',inplace=True)
data.head()
```

```python
data.head()
```

```python
data['host_since'] = pd.to_datetime(data['host_since'])

data.head()
```

```python
import datetime

latest = datetime.datetime(2023,9,5)
data['time_on_platform'] =  latest - data['host_since']

data['time_on_platform'] =data['time_on_platform'] / np.timedelta64(1,'Y')

data['time_on_platform'] = data['time_on_platform'].round(0)
```

```python
data
```

```python
data['host_listings_count'] = data['host_listings_count'].astype('int64',copy=False)
data['host_total_listings_count'] = data['host_total_listings_count'].astype('int64',copy=False)
data['time_on_platform'] = data['time_on_platform'].astype('int64',copy=False)
```

```python
data.head()
```

```python
def rem_sign(x):
    x = x.split('%')[0]
    return x
```

```python
data['host_response_rate'] = data['host_response_rate'].apply(rem_sign)
data['host_acceptance_rate'] = data['host_acceptance_rate'].apply(rem_sign)
```

```python
def rem_sign2(x):
    x = x.split('$')[1]
    return x
```

```python
data['price'] = data['price'].apply(rem_sign2)
```

```python
data.head()
```

```python
data.drop(['host_since'],axis=1,inplace=True)
```

```python
data.to_csv('hotel_aggregator.csv', sep=',', index=False, encoding='utf-8')
```

```python
data.dtypes
```

```python
def rem_comma(x):
    try:
        x = x.split(',')[0] + x.split(',')[1]
        return x
        
    except IndexError:
        return x
```

```python
data['host_response_rate'] = data['host_response_rate'].astype('int64',copy=False)
data['host_acceptance_rate'] = data['host_acceptance_rate'].astype('int64',copy=False)


data['price'] = data['price'].apply(rem_comma)
data['price'] = data['price'].astype('float64',copy=False)
```
